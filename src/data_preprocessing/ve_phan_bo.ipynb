```python
import matplotlib.pyplot as plt
import numpy as np

label_colors = ['blue', 'green', 'red', 'purple', 'orange']
# m s lng mu cho tng nhn trong tp hun luyn
unique_labels, label_counts = np.unique(uncleaned_y_train, return_counts=True)
# To biu  chng ln nhau cho tng nhn

plt.figure(figsize=(10, 6))
plt.rc('font', size=20) 

for label, color in zip(unique_labels, label_colors):
    label_data = dirty_X_train[dirty_y_train == label]  # Ly d liu ca tng nhn
    plt.hist(label_data.flatten(), color="#000000", bins=6, alpha=0.9, label='Label errors',stacked=False)

for label, color in zip(unique_labels, label_colors):
    label_data = uncleaned_X_train[uncleaned_y_train == label]  # Ly d liu ca tng nhn
    plt.hist(label_data.flatten(), color=color, bins=6, alpha=0.5, label=label)

plt.xticks(np.arange(0, 0.04, step=0.004))
plt.xlabel('Current Values (A)')
plt.ylabel('Frequency (Hz)')
# plt.title('Phn b d liu trong tng nhn cha lam sach')
# plt.legend(loc='best')
plt.axhline(0, color='black', linewidth=3)  # V trc tung
plt.axvline(0, color='black', linewidth=3)  # V trc honh
plt.grid(color='black', linestyle='--', linewidth=0.5,alpha=0.2)
plt.show()
```
