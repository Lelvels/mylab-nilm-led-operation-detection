```python
data_repo = DataRepository("../.env")
#Du lieu dong dien chua duoc lam sach
uncleaned_X_train, uncleaned_y_train, uncleaned_X_validation, uncleaned_y_validation, uncleaned_X_test, uncleaned_y_test = data_repo.load_current_data(clean_data=False)
#Du lieu dong dien da duoc lam sach
cleaned_X_train, cleaned_y_train, cleaned_X_validation, cleaned_y_validation, cleaned_X_test, cleaned_y_test = data_repo.load_current_data(clean_data=True)
#Du lieu dong dien bi loai bo trong qua trinh lam sach
dirty_X_train, dirty_y_train, dirty_X_validation, dirty_y_validation, dirty_X_test, dirty_y_test = data_repo.load_dirty_current_data()
```

```python
import matplotlib.pyplot as plt
import numpy as np

label_colors = ['blue', 'green', 'red', 'purple', 'orange']
# Đếm số lượng mẫu cho từng nhãn trong tập huấn luyện
unique_labels, label_counts = np.unique(uncleaned_y_train, return_counts=True)
# Tạo biểu đồ chồng lên nhau cho từng nhãn

plt.figure(figsize=(10, 6))
plt.rc('font', size=20) 

for label, color in zip(unique_labels, label_colors):
    label_data = dirty_X_train[dirty_y_train == label]  # Lấy dữ liệu của từng nhãn
    plt.hist(label_data.flatten(), color="#000000", bins=6, alpha=0.9, label='Label errors',stacked=False)

for label, color in zip(unique_labels, label_colors):
    label_data = uncleaned_X_train[uncleaned_y_train == label]  # Lấy dữ liệu của từng nhãn
    plt.hist(label_data.flatten(), color=color, bins=6, alpha=0.5, label=label)

plt.xticks(np.arange(0, 0.04, step=0.004))
plt.xlabel('Current Values (A)')
plt.ylabel('Frequency (Hz)')
# plt.title('Phân bố dữ liệu trong từng nhãn chưa làm sạch')
# plt.legend(loc='best')
plt.axhline(0, color='black', linewidth=3)  # Vẽ trục tung
plt.axvline(0, color='black', linewidth=3)  # Vẽ trục hoành
plt.grid(color='black', linestyle='--', linewidth=0.5,alpha=0.2)
plt.show()
```
